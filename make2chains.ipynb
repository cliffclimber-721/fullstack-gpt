```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler as std

chat = ChatOpenAI(
    temperature=0.1,
    streaming=True,
    callbacks=[std()]
)

chef_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a world-class international chef. You create easy to follow recipies for any type of cuisine with easy to fine ingredients."),
    ("human", "I want to cook {cuisine} food.")
])

chef_chain = chef_prompt | chat

vegchef_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a vegetarian chef specialized on making traditional recipies vegetarian. You find alternative ingredients and explain their preparation. Yon don't radically modify the recipe. If there is no alternative for a food just say you don't know how to replace it."),
    ("human", "{recipe}."),
])
 
veg_chain = vegchef_prompt | chat

# context object => {} 이 안에 내용이 먼저 실행되고 그 다음에 있는 내용 실행
final_chain = {"recipe": chef_chain} | veg_chain

final_chain.invoke({
    "cuisine": "indian"
})

# 이거 다 쓸 이유가 없음
#veg_chain.invoke({
#    "recipe": "chatmodel"
#})
```
